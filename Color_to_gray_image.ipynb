```python
import cv2
from google.colab.patches import cv2_imshow
imagen = cv2.imread('flower.jpg')
img_gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)

```

```python
img_converted = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
cv2_imshow(img_converted)
```

```python
b = img_converted.copy()
b[:, :, 1] = 0
b[:, :, 2] = 0
cv2_imshow(b)
```

```python
g = img_converted.copy()
g[:, :, 0] = 0
g[:, :, 2] = 0
cv2_imshow(g)
```

```python
r = img_converted.copy()
r[:, :, 0] = 0
r[:, :, 1] = 0
cv2_imshow(r)
```
